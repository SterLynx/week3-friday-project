```python
class InvestmentCalc:
    def __init__(self):
        self.total_income = 0 
        self.total_expenses = 0
        self.total_investments = 0
        self.monthly_cash_flow = 0
        self.annual_cash_flow = 0

    def get_income(self):
        rental_income = int(input("What is your monthly rental income? "))
        laundry_income = int(input("What is your monthly laundry income? "))
        storage_income = int(input("What is your monthly storage income? "))
        misc_income = int(input("What is your monthly miscellaneous income? "))

        self.total_income = rental_income + laundry_income + storage_income + misc_income
        print(f'Your total monthly income is: {self.total_income}')

    def get_expenses(self):
        monthly_tax = int(input("What is your monthly tax expense? "))
        insurance_expense = int(input("What is your monthly insurance expense? "))
        electricity_expense = int(input("What is your monthly electricity expense? "))
        water_expense = int(input("What is your monthly water expense? "))
        sewer_expense = int(input("What is your monthly sewer expense? "))
        garbage_expense = int(input("What is your monthly garbage expense? "))
        gas_expense = int(input("What is your monthly gas expense? "))
        hoa_fees = int(input("What is your monthly HOA expense? "))
        lawn_snow_fees = int(input("What is your monthly lawn and snow expense? "))
        vacancy_expense = int(input("What is your monthly vacancy expense? "))
        repair_expense = int(input("What is your monthly repair expense? "))
        capital_expend = int(input("What is your monthly capital expenditure expense? "))
        property_management = int(input("What is your monthly property management expense? "))
        mortgage_expense = int(input("What is your monthly mortgage expense? "))

        self.total_expenses = (
            monthly_tax + insurance_expense + electricity_expense +
            water_expense + sewer_expense + garbage_expense + gas_expense +
            hoa_fees + lawn_snow_fees + vacancy_expense + repair_expense +
            capital_expend + property_management + mortgage_expense
        )
        print(f'Your total monthly expenses are: {self.total_expenses}')

    def calculate_cash_flow(self):
        self.monthly_cash_flow = self.total_income - self.total_expenses
        self.annual_cash_flow = self.monthly_cash_flow * 12
        print(f'Your monthly cash flow is {self.monthly_cash_flow} and your annual cash flow is {self.annual_cash_flow}')

    def get_investments(self):
        print("Let's take a look at your investments:\n")
        down_payment = int(input("What is your down payment? "))
        closing_cost = int(input("What is your closing cost? "))
        rehab_budget = int(input("What is your rehab budget? "))
        misc_invest = int(input("How much are your miscellaneous investments? "))

        self.total_investments = down_payment + closing_cost + rehab_budget + misc_invest
        print(f'Your total investment cost is {self.total_investments}')

    def calculate_cash_on_cash(self):
        cash_on_cash_roi = (self.annual_cash_flow / self.total_investments) * 100
        print(f"Your cash on cash ROI is {cash_on_cash_roi:.2f}%")

    def run_program(self):
        self.get_income()
        self.get_expenses()
        self.calculate_cash_flow()
        self.get_investments()
        self.calculate_cash_on_cash()

if __name__ == '__main__':
    investment_calculator = InvestmentCalc()
    investment_calculator.run_program()
```
